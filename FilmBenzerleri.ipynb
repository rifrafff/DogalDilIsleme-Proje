```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import svm
from sklearn.metrics import classification_report, accuracy_score
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_excel("film_Datasi.xlsx")
df.head()
```

```python
target_column = 'Konu'
stop_words = set(stopwords.words('turkish'))
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def remove_stopwords(text):
    if isinstance(text, str):
        stop_words = set(stopwords.words('english'))
        words = word_tokenize(text)
        filtered_words = [word for word in words if word.lower() not in stop_words]
        return ' '.join(filtered_words)
    else:
        return text
```

```python
new_column = 'temizlenmis_konu'
df[new_column] = df[target_column].apply(remove_stopwords)
```

```python
df.head()
```

```python
from simplemma import text_lemmatizer

def clean_text(text):
    if isinstance(text, str):
        text = text.replace("Â", "a")
        text = text.replace("â", "a")
        text = text.replace("î", "i")
        text = text.replace("Î", "ı")
        text = text.replace("İ", "i")
        text = text.replace("I", "ı")
        text = text.replace(u"\u00A0", " ")
        text = text.replace("|", " ")

        text = re.sub(r"@[A-Za-z0-9]+", " ", text)
        text = re.sub(r"(.)\1+", r"\1\1", text)
        text = re.sub(r"https?:\/\/\S+", " ", text)
        text = re.sub(r"http?:\/\/\S+", " ", text)
        text = re.sub(r"\n", " ", text)
        text = re.sub(r"#(\w+)", " ", text)
        text = re.sub(r"^\x00-\x7F]+", " ", text)
        text = re.sub(r"[^A-Za-zâîığüşöçİĞÜŞÖÇ]+", " ", text)
        text = re.sub(r"((https://[^\s]+))", " ", text)
        text = " ".join(text.lower().strip().split())

        return " ".join(text_lemmatizer(text, lang="tr"))
    else:
        return text
```

```python
def clean_row(row):
    konu = clean_text(row["temizlenmis_konu"])
    return konu
```

```python
df["clean"] = df.apply(lambda row: clean_text(row["temizlenmis_konu"]), axis=1)
df.head()
```

```python
# Kategorileri belirleyin (örneğin: "aksiyon", "komedi", "drama" vb.)
categories = df["Türler"].unique()

# Hangi film hangi kategoriye ait olduğunu etiketleyin
def etiketle(row):
    for key, category in enumerate(categories):
        if row["Türler"] == category:
            return key

df["labels"] = df.apply(lambda row: etiketle(row), axis=1)
```

```python
# Veriyi eğitim ve test setlerine bölün
X_train, X_test, y_train, y_test = train_test_split(df["clean"], df["labels"], test_size=0.2, random_state=42)
```

```python
# X_train ve X_test'i 2D dizilere dönüştürme
X_train_2d = np.array(X_train).reshape(-1, 1)
X_test_2d = np.array(X_test).reshape(-1, 1)

# Eksik değerleri temizleme
imputer = SimpleImputer(strategy='constant', fill_value='')  # Varsayılan olarak eksik değerleri boş bir string ile dolduruyoruz.
X_train_imputed = imputer.fit_transform(X_train_2d)
X_test_imputed = imputer.transform(X_test_2d)

# TfidfVectorizer kullanarak metin verilerini vektörlere dönüştürme
vectorizer = TfidfVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train_imputed.flatten())  # flatten ile 2D diziyi tekrar 1D'ye dönüştürüyoruz
X_test_vectorized = vectorizer.transform(X_test_imputed.flatten())
```

```python
# Multinomial Naive Bayes modelini eğitin
model_NB = MultinomialNB()
model_NB.fit(X_train_vectorized, y_train)

# Modelin performansını değerlendirin
predictions_nb = model_NB.predict(X_test_vectorized)

# Benzersiz sınıfları alın
unique_classes = np.unique(predictions_nb)

# Modelin performans raporunu görüntüleme
print("Naive Bayes Model\n")
print("Accuracy:", accuracy_score(y_test, predictions_nb))
print("\nClassification Report:\n", classification_report(y_test, predictions_nb, labels=unique_classes))
```

```python
# Support Vector Machine modelini eğitin
model_SVM = svm.SVC(kernel='linear', random_state=0, probability=True)
model_SVM.fit(X_train_vectorized, y_train)

# Modelin performansını değerlendirin
predictions_svm = model_SVM.predict(X_test_vectorized)

# Benzersiz sınıfları ve sınıf etiketlerini alın
unique_classes_svm = np.unique(np.concatenate((y_test, predictions_svm))).astype(str)
labels = list(range(len(unique_classes_svm)))

# Modelin performans raporunu görüntüleme
print("\nSupport Vector Machine Model\n")
print("Accuracy:", accuracy_score(y_test, predictions_svm))
print("\nClassification Report:\n", classification_report(y_test, predictions_svm, labels=labels, target_names=unique_classes_svm))

```

```python
def build_markov_model(corpus, order=1):
    markov_model = defaultdict(list)

    # Corpus üzerinde dolaşılarak Markov modeli oluşturuluyor
    for i in range(len(corpus) - order):
        current_state = tuple(corpus[i:i + order])
        next_state = corpus[i + order]
        markov_model[current_state].append(next_state)

    return markov_model
```
